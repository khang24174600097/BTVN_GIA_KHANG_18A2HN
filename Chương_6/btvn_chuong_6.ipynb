# I. Câu hỏi thảo luận

### 1. Multithreading trong Python:
Là kỹ thuật cho phép chương trình chạy nhiều luồng (thread) song song trong cùng một tiến trình.

### 2. Giải thích khái niệm Thread trong Python:
Thread là đơn vị thực thi nhỏ nhất trong một tiến trình, dùng chung bộ nhớ với các thread khác.

### 3. Có bao nhiêu Thread chạy cùng lúc trong Python:
Có thể tạo nhiều thread, nhưng với CPU-bound thì chỉ 1 thread chạy tại một thời điểm do GIL.

### 4. Lợi ích đa luồng trong Python:
Tăng hiệu quả với tác vụ I/O, cải thiện độ phản hồi, tận dụng thời gian chờ


### 5. Sự khác nhau giữa tiến trình và luồng:
Tiến trình có bộ nhớ riêng, luồng dùng chung bộ nhớ và nhẹ hơn.

### 6. Cần phải locks trong khi viết code đa luồng: 
Để tránh xung đột dữ liệu khi nhiều luồng truy cập tài nguyên chung.

### 7. Đồng bộ hóa khi nói đến đa luồng:
Là cách kiểm soát thứ tự và thời điểm các luồng truy cập tài nguyên chung.

### 8. Các cách phổ biến để triển khai khóa trong Python:
Lock, RLock, Semaphore, Condition

### 9. Điều gì xảy ra nếu 2 luồng lấy cùng 1 khóa:
Một luồng được khóa, luồng còn lại phải chờ

### 10. Hạn chế của mô - đun threading trong Python:
Bị giới hạn bởi GIL, không hiệu quả với tác vụ tính toán nặng (CPU-bound).

# II. Bài tập vận dụng

### Bài 6.1.

```python
import threading
import time

def worker():
    print(f"Đang chạy luồng: {threading.current_thread().name}")
    time.sleep(1)

threads = []
for i in range(5):
    t = threading.Thread(target=worker, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tất cả luồng đã hoàn thành.")
```

### Bài 6.10.

```python
import threading
import random

def find_max(sub_list, results, index):
    sub_max = max(sub_list)
    results[index] = sub_max
    print(f"{threading.current_thread().name} tìm max = {sub_max}")

n = 20  
data = [random.randint(0, 100) for _ in range(n)]
print("List ban đầu:", data)

num_threads = 4
chunk_size = n // num_threads

threads = []
results = [0] * num_threads

for i in range(num_threads):
    start = i * chunk_size
    end = n if i == num_threads - 1 else (i + 1) * chunk_size
    sub_list = data[start:end]

    t = threading.Thread(
        target=find_max,
        args=(sub_list, results, i),
        name=f"Thread-{i+1}"
    )
    threads.append(t)
    t.start()

for t in threads:
    t.join()

max_value = max(results)
print("Giá trị lớn nhất trong list:", max_value)








```
