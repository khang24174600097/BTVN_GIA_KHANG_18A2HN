I - Câu hỏi lý thuyết 

Câu 1: Khái niệm về XML? 
- XML (eXtensible Markup Language) là ngôn ngữ đánh dấu mở rộng dùng để lưu trữ và trao đổi dữ liệu.
    + XML không hiển thị dữ liệu như HTML mà tập trung mô tả ý nghĩa và cấu trúc của dữ liệu.
    + Người dùng có thể tự định nghĩa các thẻ (tag) để phù hợp với nhu cầu riêng.
→ Nói ngắn gọn: XML giúp dữ liệu có cấu trúc rõ ràng, dễ chia sẻ giữa các hệ thống khác nhau.

Câu 2: Khái niệm DOM ?
- DOM (Document Object Model) là mô hình đối tượng tài liệu – biểu diễn cấu trúc của tài liệu XML (hoặc HTML) dưới dạng cây (tree).
    + Mỗi thành phần trong tài liệu (phần tử, thuộc tính, văn bản...) là một nút (node) trong cây.
    + DOM cho phép chương trình truy cập, thay đổi, thêm hoặc xóa nội dung XML bằng các ngôn ngữ lập trình như Python, JavaScript, Java,…

Câu 3: Cú pháp tài liệu XML?
- Cú pháp : 
    + <?xml version="1.0" encoding = "UTF-8">
    

Câu 4: API XML là gì?
- API XML (Application Programming Interface for XML) là giao diện lập trình ứng dụng giúp lập trình viên thao tác với dữ liệu XML.
- API cung cấp các hàm, lớp, phương thức để đọc, ghi, phân tích (parse), và chỉnh sửa XML.
- Một số API XML phổ biến:
    + DOM API: truy cập XML dưới dạng cây.
    + SAX API: xử lý XML theo dòng (sự kiện).
